```python
import pandas as pd
```

```python
df_assemblies_01 = pd.read_csv('assemblies_0111-1511.csv', sep=',')
df_assemblies_01
```

```python
df_histories_01 = pd.read_csv('histories_0111-1511.csv', sep=',')
df_histories_01
```

```python
df_histories_01['assembly_datetime'] = [
    df_histories_01['created_at'][i] if df_histories_01['status'][i] == 'assembly' else 0
    for i in range(len(df_histories_01)) 
]
df_histories_01
```

```python
df_histories_01['assembled_datetime'] = [
    df_histories_01['created_at'][i] if df_histories_01['status'][i] == 'assembled' else 0
    for i in range(len(df_histories_01)) 
]
df_histories_01
```

```python
df_hist_assembly_01 = df_histories_01[df_histories_01['assembly_datetime'] != 0]
assembly_cols = df_hist_assembly_01.columns.to_list()
assembly_cols.pop(1)
assembly_cols.pop(-2)
df_hist_assembly_01.drop(assembly_cols, axis=1, inplace=True)
df_hist_assembly_01
```

```python
df_hist_assembled_01 = df_histories_01[df_histories_01['assembled_datetime'] != 0]
assembled_cols = df_hist_assembled_01.columns.to_list()
assembled_cols.pop(1)
assembled_cols.pop(-1)
df_hist_assembled_01.drop(assembled_cols, axis=1, inplace=True)
df_hist_assembled_01
```

```python
df_histories_01 = pd.merge(df_hist_assembly_01, df_hist_assembled_01, how='inner', on='assembly_id')
df_01 = pd.merge(df_assemblies_01, df_histories_01, how='inner', on='assembly_id')
df_01
```

```python
df_01.info()
```
